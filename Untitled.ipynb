```python
def get_details(html,department=None,lab=None):
    ## the next function get html page and return the member 
    url=html
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    my_div = soup.find_all('div', {'class': 'block'})[1:] 
    # data base of the deatails
    my_div
    title_name=[]
    role=[]
    email=[]
    phone=[]
    department1=[]
    department2=[]
    #running for the first loop- the each role on the deaprtment 
    for sub_div in my_div:
        try:
            general_role=sub_div.find('h2').text
        except:
            print(sub_div)
            general_role='None'
        sub_div=sub_div.find_all('div', {'class': 'ce list-item'})
    #running for the second sub loop - each person
        for i in range(len(sub_div)):
            print(i)
            role.append(general_role)
            try:
                title_name.append(sub_div[i].find('span', {'class': 'family-name'}).text)
            except:
                print('error title',i)
                title_name.append('nan')
#                 role.append(sub_div[i].find('em'))
            try:
                span_id=sub_div[i].find_all('span')[1].get('id')
                email.append(sub_div[i].find('span', {'id': span_id}).text)
            except:
                print('error email',i)
                email.append('nan')
            try:
                phone.append(sub_div[i].find_all('p', {'class': 'tel'})[0].text)
            except:
                print('error phone',i)
                phone.append('nan')

    return title_name,role,email,phone
```

```python
data_base=pd.DataFrame(columns=['Website-link','Department','Lab/Group Name','name','Position/Function'
                               'email','phone number'])
for i in range(len(data)):
    print(i)
    title_name,role,email,phone=get_details(data.at[i,'link'])
    Website_link=[]
    department=[]
    lab=[]
    for j in range(len(title_name)):
        Website_link.append(data.at[i,'link'])
        department.append(data.at[i,'department'])
        lab.append(data.at[i,'lab'])
    new_data={'Website-link':Website_link,'Department':department,
              'Lab/Group Name':lab,'name':title_name,'Position/Function':role,
              'email':email,'phone number':phone}
    new_df = pd.DataFrame(new_data)

    # append the new DataFrame to the original DataFrame
    data_base = data_base.append(new_df, ignore_index=True)
```
